```python
journeys = []

def replaceitem(x):
    if abs(x) > 20:
        return np.random.randint(2, 21)
    elif abs(x) > 1:
        return int(abs(x))
    else:
        return np.random.randint(2, 21)

for i in range(10000):
    n = int(round(abs(np.random.normal(4, 4, 1)[0]))) 
    n = n if n > 1 else 4
    path = np.random.normal(4, 4, n)
    path = [i for i in path]
    path = map(replaceitem, path)
    #print(list(path))
    journeys.append(path)
```

```python
df = pd.DataFrame(journeys)
```

```python
df.describe()
```

```python
df.head()
```

```python
df['Path'] = df[df.columns].apply(
    lambda x: ' > '.join(x.dropna().astype(str).str[:-2]),
    axis=1)
```

```python
def conversion_conditions(x):
    if '7' in x.split(' > '):
        return 1
    elif ('6' in x.split(' > ')) and\
         ('3' in x.split(' > ')):
        return 1
    else:
        return 0

df['Conversion'] = df['Path'].apply(lambda x: conversion_conditions(x))
```

```python
df['Conversion'].value_counts(normalize = True, dropna = False)
```

```python
df['Conversion Value'] = df['Conversion'].apply(lambda x: x * abs(np.random.normal(200, 50, 1)[0]))
```

```python
df = df[['Path', 'Conversion', 'Conversion Value']]
```

```python
df = df.groupby('Path').sum().sort_values(by = 'Conversion', ascending = False).reset_index()
```

```python
df.to_csv('Paths.csv', index = False)
```
